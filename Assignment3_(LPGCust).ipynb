```python
# 1.a)  Displaying number of LPG customers in 2020 Region-wise as bar-graph

state = ds['State']
statetotal = ds['Total2020']
 
# Figure Size
fig = plt.figure(figsize =(100, 7))
 
# Horizontal Bar Plot
plt.bar(state[:], statetotal[:])
 
# Show Plot
plt.show()
```

```python
# 1.b)  Displaying number of LPG customers in 2020 Region-wise as pie-chart

state = ds['State']
statetotal = ds['Total2020']

plt.pie(statetotal, shadow = True)
plt.legend(state, loc="center right", bbox_to_anchor=(1.5, 0, 0.5, 1))

plt.show() 
```

```python
# 2. Displaying Top 5 states in 2020
ds2 = ds.sort_values("Total2020", ascending=False)
ds3 = ds2.head(5)

state = ds3['State']
statetotal = ds3['Total2020']
 
# Figure Size
fig = plt.figure(figsize =(10, 7))
 
# Horizontal Bar Plot
plt.bar(state, statetotal)
 
# Show Plot
plt.show()
```

```python
# 3. Displaying bottom 5 states in 2020

ds2 = ds.sort_values("Total2020", ascending=False)
ds4 = ds2.tail(5)

state = ds5['State']
statetotal = ds5['Total2020']
 
# Figure Size
fig = plt.figure(figsize =(20, 7))
 
# Horizontal Bar Plot
plt.bar(state, statetotal)
 
# Show Plot
plt.show()
```

```python
# Displaying company-wise shares in LPG connections with pie-chart

# Calculating total shares
IOCL_total = ds["IOCL"].sum()
HPCL_total = ds["HPCL"].sum()
BPCL_total = ds["BPCL"].sum()

# Displaying PieChart
y = np.array([IOCL_total, HPCL_total, BPCL_total])
mylabels = ["IOCL", "HPCL", "BPCL"]
myexplode = [0.05, 0.05, 0.05]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
state = ds['State']
total2020 = ds['Total2020']
total2019 = ds['Total2019']

n=len(ds)
r = np.arange(n)
width = 0.25
  
fig = plt.figure(figsize =(100, 7))

bar2019 = plt.bar(r, total2019, color = 'b', width = width, edgecolor = 'black', label='2019')
bar2020 = plt.bar(r + width, total2020, color = 'g', width = width, edgecolor = 'black', label='2020')

plt.xticks(r, state)

plt.legend()

plt.show()
```
